## Tipos de datos

```python
# enter (Int)
edad = 18
type(edad) # me dice el tipo de dato

# Decimales (flast)
altura = 1.67
type(altura)

# Cadenas de texto (str)
nombre = "Ursula"
type(nombre)

#Booleanos (bool)
activo = True # False
type(activo)

#Los prints son mas frecuentes en los notebooks
print("Tipo de edad:", type(edad))
print("Tipo de altura:", type(altura))
print("tipo de nombr:", type(nombre))
print("Tipo de activo:", type(activo))
```

Se introduce la funcion format()

```python
usuario = "Ursula"
mensaje = "Como te puedo ayudar?"

#format() permite insertar valores a un print por medio de dos llaves
print("Hola {}! {}". format(usuario, mensaje))
```
